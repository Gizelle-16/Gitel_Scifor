```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Employee_Salary_Dataset.csv')
df.info

```

```python
#bar chart
x = df['Experience_Years']
y = df['Salary']

plt.bar(x,y, color ='pink')
```

```python
#piechart
y = df['Experience_Years']
fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(x, autopct='%.1f%%')
ax.set_title('PIe chart salary')
plt.tight_layout()
```

```python
#histogram
plt.figure(figsize = (8,5))
x = df["Age"]

plt.hist(x, bins = 20, color = "purple")
plt.title("Employee ages")
plt.xlabel("Age")
plt.ylabel("Count")
```

```python
#boxplot
new_data = df[["Experience_Years"]]
print(new_data.head())
plt.figure(figsize = (10, 7))
new_data.boxplot()
```

```python
#scatterplot
x =df['Experience_Years']
y =df['Salary']
plt.scatter(x,y)
```
